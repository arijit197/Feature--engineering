## **Missing data not at random (MNAR)**

**(percentage of null values is greater than 90%)**

```python
#MiscFeature(96),PoolQC(99),Alley(93)
i = 0
ar = []
for a in y:
    if a >= 0.9  :
        print(y.index[i],' : ',a)
        ar.append(y.index[i])
    i+=1
```
